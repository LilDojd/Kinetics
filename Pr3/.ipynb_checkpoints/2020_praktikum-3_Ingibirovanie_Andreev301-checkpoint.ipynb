```python
# Get data into dataframe
import pandas as pd
from itertools import product
import numpy as np
import altair as alt
from vega_datasets import data
read_df = pd.read_excel('2020_практикум-3_301_Андреев.xls', header=[0,1,2], na_values=['--'])[:300]
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
# Приведем в нужный вид
df = pd.DataFrame(columns=['time', 'O2 depletion', 'Experiment'])
for i, (SI, df_slice) in enumerate(read_df.groupby(level=[0, 1], axis=1)):
    new_df = df_slice.copy()
    S, I = SI
    new_df.columns = ['time', 'O2 depletion']
    new_df['Experiment'] = f"{S}, {I}"
    df = df.append(new_df, ignore_index = True)
df = df.sort_values(by=['time']).reset_index(drop=True)
df['time'] += 0.078
```

```python
def make_altair(experiment):
    
    slider = alt.binding_range(min=0, max=100, step=0.5, name='Linear region:')
    selector = alt.selection_single(name="LR", fields=['cutoff'],
                                bind=slider, init={'cutoff': 10})
    
    base = alt.Chart(data[data['Experiment'] == experiment]).mark_circle(opacity=0.3).encode(
        x = alt.X('time:Q', axis = alt.Axis(title = 'Time, s')),
        y = alt.Y('O2 depletion:Q',
                  axis = alt.Axis(title = 'O2 depletion, ')
                 ),
        color=alt.condition(
            alt.datum.time < selector.cutoff,
            alt.value('red'), alt.value('blue'))
    ).properties(
        title = f'{experiment}',
        width=600,
        height=300
    )
    
    red_line = base.transform_filter(alt.datum.time < selector.cutoff).mark_line().encode(color=alt.value('red'))
    blue_line = base.transform_filter(alt.datum.time >= selector.cutoff).mark_line().encode(color=alt.value('blue'))
    
    polynomial_fit = red_line.transform_regression(
        "time",
        "O2 depletion",
        extent=[0, max(data[data['Experiment'] == experiment].dropna()['time'])]
    ).mark_line(clip=True).encode(
        color=alt.value('red'),
        y = alt.Y('O2 depletion:Q', 
                  scale=alt.Scale(
                      domain=[0, max(data[data['Experiment'] == experiment].dropna()['O2 depletion'])])
                 )
    )
    
    fit_coef = red_line.transform_regression(
        'time',
        'O2 depletion',
        params=True
    ).mark_text(align='left', size=20).encode(
        x=alt.value(400),  # pixels from left
        y=alt.value(250),  # pixels from top
        text='label:N',
        color=alt.value('black')
    ).transform_calculate(label='"y = " + format(datum.coef[0],".3f") + " + " + format(datum.coef[1],".3f") + "x"')
    
    initial_speed = red_line.transform_regression(
        'time',
        'O2 depletion',
        params=True
    ).mark_text(align='left', size=20).encode(
        x=alt.value(350),  # pixels from left
        y=alt.value(275),  # pixels from top
        text='label:N',
        color=alt.value('black')
    ).transform_calculate(label='"V (по субстрату) = " + format(datum.coef[1]/2,".3f")')

    chart = base.add_selection(selector) + red_line + blue_line + polynomial_fit + fit_coef + initial_speed
    
    return chart.interactive()
```

```python
# Копия таблицы из pdf
rates = {0: [0.447, 0.496, 0.805, 1.008],
         0.07: [0.257, 0.279, 0.454, 0.572],
         0.2: [0.135, 0.158, 0.244, 0.303]}
SI_df = pd.DataFrame.from_dict(rates, columns=[5.0, 6.66, 20.0, 100.0],
                               orient='index')
SI_df = SI_df.T
SI_df.index.name = 'Substrate_concentration'
SI_df = SI_df.reset_index()
long_SI = SI_df.melt('Substrate_concentration', var_name='Inhibitor concentration', value_name='Vinitial')
long_SI
```

## Здесь можно навести мышку на линию, и высветятся ее параметры

```python
long_SI['Fitted_line'] = ['1/V = 0.925 + 6.798/S'] * 4 + ['1/V = 1.647 + 11.797/S'] * 4 + ['1/V = 3.052 + 21.776/S'] * 4
base = alt.Chart(long_SI).mark_circle(size=60).encode(
    x = alt.X('x:Q', axis=alt.Axis(title = '1/S, 1/μM')),
    y = alt.Y('y:Q', axis=alt.Axis(title = '1/V, s/μM')),
    color='Inhibitor concentration:N',
).transform_calculate(
    x='1/datum.Substrate_concentration',
    y='1/datum.Vinitial'
).properties(
        title = 'Lineweaver–Burk plot',
        width=600,
        height=600
)

# Create a selection that chooses the nearest point & selects based on x-value
nearest = alt.selection(type='single', nearest=True, on='mouseover',
                        fields=['x'], empty='none')

    
fit = base.transform_regression("x",
                                "y",
                                groupby=['Inhibitor concentration'],
                                extent=[-0.18, max(1/long_SI['Substrate_concentration'].values) + 0.2],
                               ).mark_line(clip=True, opacity=0.5)
locs = [200, 260, 320]
names = [0, 0.07, 0.2]

# fit_t = [alt.Chart(long_SI[long_SI['Inhibitor concentration'] == name]).transform_calculate(    x='1/datum.Substrate_concentration',
#     y='1/datum.Vinitial').transform_regression("x",
#                                   "y",
#                                   groupby=['Inhibitor concentration'],
#                                   params=True
#                                  ).mark_text(align='left', size=20).encode(
#         x=alt.value(400),  # pixels from left
#         y=alt.value(loc),  # pixels from top
#         text='label:N',
#         color=alt.value('black')
#     ).transform_calculate(label='"y = " + format(datum.coef[0],".3f") + " + " + format(datum.coef[1],".3f") + "x"') for loc, name in zip(locs, names)]

# Transparent selectors across the chart. This is what tells us
# the x-value of the cursor
selectors = base.mark_point().encode(
    x='x:Q',
    opacity=alt.value(0),
).add_selection(
    nearest
)
    
# Draw text labels near the points, and highlight based on selection
text = base.mark_text(align='left', dx=-100, dy=-20, size=15).encode(
    text=alt.condition(nearest, 'Fitted_line:N', alt.value(' '))
)

# Draw a rule at the location of the selection
rules = base.mark_rule(size=1.5).encode(
    x='x:Q',
    color=alt.value('#d3d3d3')
).transform_filter(
    nearest
)

y_coords = alt.Chart().mark_rule().encode(x = 'a:Q', color = alt.value('black')).transform_calculate(a="0")
x_coords = alt.Chart().mark_rule().encode(y = 'a:Q', color = alt.value('black')).transform_calculate(a="0")


chart = (base + fit + x_coords + y_coords + rules + text + selectors).configure_axis(
    labelFontSize=20,
    titleFontSize=20).interactive()
chart
```

### Очень низкое сходство
#### Построим график 1/Vm* (I) 
$$
\frac{1}{V_{m}^{*}}=\frac{1+\frac{I}{K_{i}}}{V_{m}}
$$, где $$
V_{m}=\frac{1}{A}, \quad K_{i}=\frac{A}{B}
$$

```python
Vm_star = pd.DataFrame.from_dict({0:1.081, 0.07: 0.607, 0.2:0.328}, orient='index', columns=['Vm_star'])
Vm_star.index.name = 'I_concentration'
Vm_star.reset_index(inplace=True)
Vm_star
```

```python
chart_Vm = alt.Chart(Vm_star).mark_point().encode(
    x=alt.X('x:Q', axis=alt.Axis(title = 'I, μM')),
    y=alt.Y('y:Q', axis=alt.Axis(title = '1/Vm*, s/μM'))
).transform_calculate(
    x='datum.I_concentration',
    y='1/datum.Vm_star'
).properties(
        title = 'Vm* by Inhibitor concentration',
        width=600,
        height=600
)

fit_Vm = chart_Vm.transform_regression('x', 'y').mark_line()

fit_coef_Vm = chart_Vm.transform_regression(
    'x',
    'y',
    params=True
).mark_text(align='left', size=20).encode(
    x=alt.value(300),  # pixels from left
    y=alt.value(300),  # pixels from top
    text='label1:N',
    color=alt.value('black')
).transform_calculate(
    label1='"y = " + format(datum.coef[0],".3f") + " + " + format(datum.coef[1],".3f") + "x"'
)

VmKi = fit_coef_Vm.mark_text(align='left', size=20).encode(
    x=alt.value(300),  # pixels from left
    y=alt.value(340),  # pixels from top
    text='label2:N',
    color=alt.value('black')
).transform_calculate(
    label2='"Vm = " + format(1/datum.coef[0],".3f") + ", Ki = " + format(datum.coef[0]/datum.coef[1],".3f")'
)

(chart_Vm + fit_Vm + fit_coef_Vm + VmKi).configure_axis(
    labelFontSize=20,
    titleFontSize=20).interactive()
```
